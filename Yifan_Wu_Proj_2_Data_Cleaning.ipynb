# 2nd Shiny: Crime at Different Location

```python
df4 = pd.read_csv("Dataset4-Violent_Crime___Property_Crime_Statewide_Totals__1975_to_Present.csv")
df4.rename(columns={'M/V THEFT PER 100,000 PEOPLE':'MotorVehicle Theft', 
                   'LARCENY THEFT PER 100,000 PEOPLE':'Larceny Theft',
                   'B & E PER 100,000 PEOPLE':'B&E',
                   'AGG. ASSAULT PER 100,000 PEOPLE':'Agg. Assult',
                   'ROBBERY PER 100,000 PEOPLE':'Robbery',
                   'RAPE PER 100,000 PEOPLE':'Rape',
                   'MURDER PER 100,000 PEOPLE': 'Murder',
                   'VIOLENT CRIME PERCENT': 'vio_percent',
                   'PROPERTY CRIME PERCENT': 'prop_percent',
                   'GRAND TOTAL': 'total'}, inplace=True)

sub = df4[['Murder', 'Rape', 'Robbery', 'Agg. Assult', 'B&E', 'Larceny Theft', 'MotorVehicle Theft', 
           'vio_percent', 'prop_percent', 'total']]

sub['year'] = list(range(1975, 2017))

#sub.iloc[0]
sub.dropna(how = 'any')
sub['vio_percent'] = [float(i[:-1]) for i in sub['vio_percent']]
sub['prop_percent'] = [float(i[:-1]) for i in sub['prop_percent']]
sub
#df4.iloc[0]
```

# Data Cleaning for d3.js chart: How Unemployment Rate, the Percentage that is Lower Than High School, and the Percentage That is Below Poverty Line can affect Total Crime Number in a Coiunty in Maryland. 

```python
#county: GEO.display-label
#unemployed: HC01_EST_VC97
#less than high school: HC01_EST_VC47
#Below 100 percent of the poverty level: HC01_EST_VC118 
crime = pd.read_csv('Dataset5-Violent_Crime___Property_Crime_by_County__1975_to_Present.csv')
crime = crime[crime['YEAR'] == '01/01/2016 12:00:00 AM']
crime.loc[(crime["JURISDICTION"] == "Allegany County "), 'JURISDICTION'] = "Allegany County"
crime.loc[(crime["JURISDICTION"] == "Anne Arundel County "), 'JURISDICTION'] = "Anne Arundel County"
crime['county'] = crime['JURISDICTION']
crime = crime[['county', 'OVERALL CRIME RATE PER 100,000 PEOPLE']]

demographic = pd.read_csv('demographic.csv')[['GEO.display-label', 'HC01_EST_VC97', 'HC01_EST_VC47', 'HC01_EST_VC118']][1:]
demographic.rename(columns={'GEO.display-label':'county', 
                   'HC01_EST_VC97':'unemployed',
                   'HC01_EST_VC47':'less_than_highschool',
                   'HC01_EST_VC118':'below_poverty'}, inplace=True)
demographic['county'] = [i[:-10] for i in demographic['county']]
demographic.loc[(demographic["county"] == "Baltimore city"), 'county'] = "Baltimore City"

merged = pd.merge(crime, demographic, on = ['county', 'county'])
merged.rename(columns={'OVERALL CRIME RATE PER 100,000 PEOPLE':'crime'}, inplace=True)
merged = merged.sort_values(by=['crime'])
merged['color'] = ["#FFFFCC", '#FFFFCC', '#FFFFCC', "#FFEDA0", "#FFEDA0", "#FED976", "#FED976",
                   "#FEB24C", "#FEB24C", "#FD8D3C", "#FD8D3C", "#FC4E2A", "#FC4E2A",
                   "#E31A1C", "#FC4E2A", "#BD0026", "#BD0026", "#BD0026", "#800026", "#800026", "gray0"] 
merged.to_csv('merged.csv')
merged
```
